```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\meswa\\OneDrive\\Desktop\\new file\\second_cle.csv")

```

```python
df['CPU_name'] = df['Cpu'].apply(lambda text:" ".join(text.split()[:3]))
```

```python
def update_cpu_name(cpu):
    if 'Intel' in cpu:
        if 'Core i7' in cpu:
            return 'Intel Core i7'
        elif 'Core i5' in cpu:
            return 'Intel Core i5'
        elif 'Core i3' in cpu:
            return 'Intel Core i3'
        else:
            return 'Intel Processor'
    elif 'AMD' in cpu:
        return 'AMD Processor'
    else:
        return cpu
```

```python
df['CPU_name'] = df['CPU_name'].apply(update_cpu_name)
```
