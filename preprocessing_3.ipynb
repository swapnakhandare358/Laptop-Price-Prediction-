```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\meswa\\OneDrive\\Desktop\\new file\\third_cle.csv")
```

```python
df['Memory'] = df['Memory'].astype(str).replace('\.0','',regex = True)
```

```python
df['Memory'] = df['Memory'].str.replace('GB','')
```

```python
df['Memory'] = df['Memory'].str.replace('TB','000')
```

```python
newdf = df['Memory'].str.split("+",n = 1,expand = True)
```

```python
df['first'] = newdf[0]
df['first'] = df['first'].str.strip()
```

```python
def applychanges(value):
    
    df['Layer1'+value] = df['first'].apply(lambda x:1 if value in x else 0)
```

```python
listtoapply = ['HDD','SSD','Hybrid','Flash Storage']   
```

```python
for value in listtoapply:
    applychanges(value)
```

```python
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
df['first'] = df['first'].str.replace(r'\D', '', regex=True)
```

```python
df['first'] = pd.to_numeric(df['first'], errors='coerce')
```

```python
df['Second'] = newdf[1]
```

```python
def applychanges1(value):
    
    df['Layer2'+value] = df['Second'].apply(lambda x:1 if value in x else 0)
```

```python
listtoapply1 = ['HDD','SSD','Hybrid','Flash Storage']
```

```python
df['Second'] = df['Second'].fillna("0")
```

```python
for value in listtoapply1:
    applychanges1(value)
```

```python
df.head(2)
```

```python
df['Second'] = df['Second'].str.replace(r'\D', '', regex=True)
```

```python
df['first'] = df['first'].astype('int')
df['Second'] = df['Second'].astype('int')
```

```python
df["HDD"]=(df["first"]*df["Layer1HDD"]+df["Second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["Second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["Second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash Storage"]+df["Second"]*df["Layer2Flash Storage"])
```

```python
df.drop(columns=['first', 'Second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash Storage'],inplace=True)
```
